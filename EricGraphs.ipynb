```python
def plotfile(fn, title):
    d = np.loadtxt(fn, skiprows=14)
    plt.plot(d[:,0], d[:,1])

    plt.title(title)
    plt.xlabel('Wavelength (nm)')
    
    #plt.ylim(0, 14000)

    plt.savefig('{0}'.format(fn)[:-4]+'.png', dpi=300)
    plt.show()
```

```python
plotfile('EricGraphs/20170906_ORNL-Torch-Sil_Subt14_14-18-40-308.txt', 'Silica in Pure Ar Torch')
```

```python
plotfile('EricGraphs/20170804-ORNL-Ar-NYSim_Subt2_13-21-23-837.txt', 'Urban')
```

```python
plotfile('EricGraphs/20170807-ORNL-Trin_Subt2_13-13-23-821.txt', 'Desert')
```
